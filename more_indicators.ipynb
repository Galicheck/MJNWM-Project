```python
import pandas as pd
```

```python
df = pd.read_csv('data/data.csv')
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df.head()
```

```python
gdp = pd.read_csv('data/GDP.csv')
gdp.columns=['Date', 'GDP']
gdp['Date'] = pd.to_datetime(gdp['Date'])
gdp.head()
```

```python
def get_quarter(date):
    quarter = (date.month - 1) // 3 + 1
    return quarter
```

```python
gdp['Year'] = gdp['Date'].dt.year
gdp['Date'] = pd.to_datetime(gdp['Date'])
gdp['Quarter'] = gdp['Date'].apply(get_quarter)
df['Quarter'] = df['Date'].apply(get_quarter)
```

```python
with_gdp = pd.merge(df, gdp, on=['Year', 'Quarter'], how='left')
with_gdp.drop(columns='Date_y', inplace=True)
with_gdp = with_gdp.rename(columns={'Date_x': 'Date'})
with_gdp['Month'] = with_gdp['Date'].dt.month
with_gdp.head()
```

```python
cci = pd.read_csv('data/cci.csv', sep=';', usecols = ['Date', 'CCI Total', 'CCI USA', 'CCI Europe'])
cci['Date'] = pd.to_datetime(cci['Date'])
cci['Month'] = cci['Date'].dt.month
cci['Year'] = cci['Date'].dt.year
cci.dropna(inplace=True)
cci.head()
```

```python
with_cci = pd.merge(with_gdp, cci, on=['Year', 'Month'], how='left')
with_cci.drop(columns='Date_y', inplace=True)
with_cci = with_cci.rename(columns={'Date_x': 'Date'})
with_cci.head()
```

```python
unemp = pd.read_csv('data/unempl_USA.csv', sep=';', usecols=['Date', 'Unemp_rate USA'])
unemp['Date'] = pd.to_datetime(unemp['Date'])
unemp['Month'] = unemp['Date'].dt.month
unemp['Year'] = unemp['Date'].dt.year
unemp.head()
```

```python
final_df = pd.merge(with_cci, unemp, on=['Year', 'Month'], how='left')
final_df.drop(columns='Date_y', inplace=True)
final_df = final_df.rename(columns={'Date_x': 'Date'})
final_df.drop(columns=['Month', 'Quarter', 'Year'], inplace = True)
final_df = final_df.rename(columns={'Unemp_rate USA': 'Unemployment_USA'})
final_df.head()
```

```python
final_df.to_csv('data/final_data.csv')
```
