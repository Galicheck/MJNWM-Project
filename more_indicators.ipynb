```python
import pandas as pd
import yfinance as yf
```

```python
df = pd.read_csv('data/data.csv')
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df.head()
```

```python
gdp = pd.read_csv('data/GDP.csv')
gdp.columns=['Date', 'GDP']
gdp['Date'] = pd.to_datetime(gdp['Date'])
gdp.head()
```

```python
def get_quarter(date):
    quarter = (date.month - 1) // 3 + 1
    return quarter
```

```python
gdp['Year'] = gdp['Date'].dt.year
gdp['Date'] = pd.to_datetime(gdp['Date'])
gdp['Quarter'] = gdp['Date'].apply(get_quarter)
df['Quarter'] = df['Date'].apply(get_quarter)
```

```python
with_gdp = pd.merge(df, gdp, on=['Year', 'Quarter'], how='left')
with_gdp.drop(columns='Date_y', inplace=True)
with_gdp = with_gdp.rename(columns={'Date_x': 'Date'})
with_gdp['Month'] = with_gdp['Date'].dt.month
with_gdp.head()
```

```python
cci = pd.read_csv('data/cci.csv', sep=';', usecols = ['Date', 'CCI Total', 'CCI USA', 'CCI Europe'])
cci['Date'] = pd.to_datetime(cci['Date'])
cci['Month'] = cci['Date'].dt.month
cci['Year'] = cci['Date'].dt.year
cci.dropna(inplace=True)
cci.head()
```

```python
with_cci = pd.merge(with_gdp, cci, on=['Year', 'Month'], how='left')
with_cci.drop(columns='Date_y', inplace=True)
with_cci = with_cci.rename(columns={'Date_x': 'Date'})
with_cci.head()
```

```python
unemp = pd.read_csv('data/unempl_USA.csv', sep=';', usecols=['Date', 'Unemp_rate USA'])
unemp['Date'] = pd.to_datetime(unemp['Date'])
unemp['Month'] = unemp['Date'].dt.month
unemp['Year'] = unemp['Date'].dt.year
unemp.head()
```

```python
not_final_df = pd.merge(with_cci, unemp, on=['Year', 'Month'], how='left')
not_final_df.drop(columns='Date_y', inplace=True)
not_final_df = not_final_df.rename(columns={'Date_x': 'Date'})
not_final_df.drop(columns=['Month', 'Quarter', 'Year'], inplace = True)
not_final_df = not_final_df.rename(columns={'Unemp_rate USA': 'Unemployment_USA'})
not_final_df.head()
```

```python
not_final_df.tail()
```

```python
start_date = '1997-05-15'
end_date = '2023-12-29'
```

```python
bac_data = yf.download('BAC', start=start_date, end=end_date)
bac_data.drop(columns=['Open', 'High', 'Low', 'Volume', 'Adj Close'], inplace=True)
bac_data.reset_index(inplace=True)
bac_data = bac_data.rename(columns={'Close': 'BAC_Close'})

jpm_data = yf.download('JPM', start=start_date, end=end_date)
jpm_data.drop(columns=['Open', 'High', 'Low', 'Volume', 'Adj Close'], inplace=True)
jpm_data.reset_index(inplace=True)
jpm_data = jpm_data.rename(columns={'Close': 'JPM_Close'})

citigroup_data = yf.download('C', start=start_date, end=end_date)
citigroup_data.drop(columns=['Open', 'High', 'Low', 'Volume', 'Adj Close'], inplace=True)
citigroup_data.reset_index(inplace=True)
citigroup_data = citigroup_data.rename(columns={'Close': 'Citigroup_Close'})

wfc_data = yf.download('WFC', start=start_date, end=end_date)
wfc_data.drop(columns=['Open', 'High', 'Low', 'Volume', 'Adj Close'], inplace=True)
wfc_data.reset_index(inplace=True)
wfc_data = wfc_data.rename(columns={'Close': 'WFC_Close'})
```

```python
d1 = pd.merge(bac_data, jpm_data, how='left', on='Date')
d2 = pd.merge(d1, citigroup_data, how='left', on='Date')
d3 = pd.merge(d2, wfc_data, how='left', on='Date')
```

```python
final_df = pd.merge(not_final_df, d3, how='left', on='Date')
```
